## 1. Use the Random Number generators discussed in the class to do the following:

### (a)Using the LGM method generate 10,000 Uniformly distributed random numbers on [0,1] and compute the empirical mean and the standard deviation of the sequence.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import norm
import pandas as pd
```

```python
a=7**5
b=0
m=(2**31)-1
x_n= random.randint(0,m-1)
LGM_lst=[x_n]
for i in range(0,9999):
    u_n1= (LGM_lst[i]*a+b)%m
    LGM_lst.append(u_n1)
uni_LGM = []
for j in range(len(LGM_lst)):
    uni_n = LGM_lst[j]/m
    uni_LGM.append(uni_n)

mean = np.mean(uni_LGM)
std = np.std(uni_LGM)
print("The empirical mean is", mean)
print("The empirical standard deviation is", std)
```

```python
plt.hist(uni_LGM,density=True)
plt.title('continuous distribution')
plt.show()
```

### (b) Use built-in functions of the software you are using to do the same thing as in (a).

```python
plt.hist(lst,density=True)
plt.title('continuous distribution')
plt.show()
```

```python
count = len([num for num in bino_sim if num >+ 40])
p_greaterthan_40 = count/len(bino_sim)
p_greaterthan_40
```

```python
from scipy.stats import binom

n = 44
p = 0.64
x = 40

# Find the probability that X is less than 40
prob_less_than_40 = binom.cdf(x-1, n, p)

# Find the probability that X is at least 40
prob_at_least_40 = 1 - prob_less_than_40

print("The probability that X is at least 40 is:", prob_at_least_40)
```

```python
def uni_sim(a,b,m,n_r,LGM):
    for i in range(0,n_r-1):
        u_n = (LGM[i]*a+b)%m
        LGM.append(u_n)
    uni_5 = []
    for j in range(len(LGM)):
        uni_n = LGM[j]/m
        uni_5.append(uni_n)
    return uni_5
```

```python
def norm_sim_bm(times,uni_dist1,uni_dist2):
    norm_sim = []
    for i in range(0,times-1):
        z_1 = np.sqrt(-2*np.log(uni_dist1[i]))*np.cos(2*np.pi*uni_dist2[i])
        z_2 = np.sqrt(-2*np.log(uni_dist1[i]))*np.sin(2*np.pi*uni_dist2[i])
        norm_sim.append(z_1+z_2)
    return norm_sim
```

```python
def norm_sim_pm(times,uni_dist1,uni_dist2):    
    norm_sim = []
    for i in range(0,times-1):
        v_1 = 2*uni_dist1[i]-1
        v_2 = 2*uni_dist2[i]-1
        w = v_1**2+v_2**2
        if w <= 1:   
            z_1 = v_1*np.sqrt(-2*np.log(w)/w)
            z_2 = v_2*np.sqrt(-2*np.log(w)/w)
            norm_sim.append(z_1+z_2)
    return norm_sim
```
